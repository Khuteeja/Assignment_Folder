```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data for males and females
np.random.seed(123)  # Set seed for reproducibility

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
male_data = {
    'Day': days,
    'Sleep_Time': np.random.randint(300, 540, size=7),  # Random sleep time for males in minutes (5 to 9 hours)
    'Study_Time': np.random.randint(60, 240, size=7),  # Random study time for males in minutes (1 to 4 hours)
    'Entertainment_Time': np.random.randint(120, 360, size=7),  # Random entertainment time for males in minutes (2 to 6 hours)
}

female_data = {
    'Day': days,
    'Sleep_Time': np.random.randint(330, 570, size=7),  # Random sleep time for females in minutes (5.5 to 9.5 hours)
    'Study_Time': np.random.randint(45, 210, size=7),  # Random study time for females in minutes (45 minutes to 3.5 hours)
    'Entertainment_Time': np.random.randint(90, 330, size=7),  # Random entertainment time for females in minutes (1.5 to 5.5 hours)
}

# Creating DataFrames for males and females
male_df = pd.DataFrame(male_data)
female_df = pd.DataFrame(female_data)

# Histogram for Sleep Time comparison between males and females
plt.figure(figsize=(8, 6))
plt.hist(male_df['Sleep_Time'], bins=10, alpha=0.5, label='Male')
plt.hist(female_df['Sleep_Time'], bins=10, alpha=0.5, label='Female')
plt.title('Sleep Time Comparison')
plt.xlabel('Sleep Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for Study Time comparison between males and females
plt.figure(figsize=(8, 6))
plt.bar(male_df['Day'], male_df['Study_Time'], label='Male', alpha=0.7)
plt.bar(female_df['Day'], female_df['Study_Time'], label='Female', alpha=0.7)
plt.title('Study Time Comparison')
plt.xlabel('Day')
plt.ylabel('Study Time (minutes)')
#plt.grid(True)
plt.show()
```

```python
# Pie chart for Entertainment Time comparison between males and females
labels = ['Male', 'Female']
sizes = [male_df['Entertainment_Time'].sum(), female_df['Entertainment_Time'].sum()]

plt.figure(figsize=(4,4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Total Entertainment Time Comparison')
plt.show()
```

```python
# Box plot for Sleep Time comparison between males and females
plt.figure(figsize=(8, 6))
plt.boxplot([male_df['Sleep_Time'], female_df['Sleep_Time']], labels=['Male', 'Female'])
plt.title('Sleep Time Comparison')
plt.ylabel('Sleep Time (minutes)')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Study Time and Entertainment Time comparison between males and females
plt.figure(figsize=(8, 6))
plt.scatter(male_df['Study_Time'], male_df['Entertainment_Time'], label='Male', color='blue')
plt.scatter(female_df['Study_Time'], female_df['Entertainment_Time'], label='Female', color='red')
plt.title('Study Time vs Entertainment Time')
plt.xlabel('Study Time (minutes)')
plt.ylabel('Entertainment Time (minutes)')
plt.grid(True)
plt.show()
```
